Before performing multivariate analysis, we have to convert our catagorical data into Numeric. we already have some columns value converted into numeric. let's import that file.

```python
# let's drop City and pincode columns for data analysis.
df = data.drop(columns=['City'],axis=1)
```

```python
# Before that we need to convert our catagorical data into numaric.
from sklearn.preprocessing import LabelEncoder
l_encoder = LabelEncoder()
df[headers] = data[headers].apply(l_encoder.fit_transform) # Applying label encoder on all the catagorical columns
```
